```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('rice.csv')
df
```

```python
from sklearn.preprocessing  import LabelEncoder
label_encoder = LabelEncoder()

country_encoded = pd.get_dummies(df,columns=['Country'])

df['Country'] = label_encoder.fit_transform(df['Country'])
```

```python
y = df['2020']
x = df.drop('2020',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

trained_mse = mean_squared_error(y_train,y_lr_train_pred)
trained_r2 = r2_score(y_train,y_lr_train_pred)

test_mse = mean_squared_error(y_test,y_lr_test_pred)
test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
results = pd.DataFrame([trained_mse,trained_r2,test_mse,test_r2]).transpose()
```
