```python
#Linear regression project for HR team to calculate the salary thats hould be given to the candidate
#based upon their experience,test_score and interview_score
```

```python
import pandas as pd
df =  pd.read_csv('hiring.csv')
df
```

```python
df.experience.fillna("zero",inplace=True)#filling the blank areas with zero as the number conversion is there
```

```python
from word2number import w2n
df.experience = df.experience.apply(w2n.word_to_num)#converting word numbers to numeric numbers
df
```

```python
median_test = df['test_score(out of 10)'].median()#finding out the median of the column
median_test
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(median_test)#setting the median in the null
```

```python
df
```

```python
import numpy as np
from sklearn import linear_model


reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
#2 year experience,9 test score , 6 interview score
#12 year experience, 10 test score , 10 interview score

reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
